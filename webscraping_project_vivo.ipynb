```python
import requests
```

```python
import bs4
```

```python
url="https://shop.vivo.com/in/products/phone"
```

```python
data=requests.get(url)
data
```

```python
data.text
type(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(data.text,"html")
```

```python
soup.find('h3',attrs={"class":"thick-font"}).text
```

```python
soup.find('div',attrs={"class":"description"}).text
```

```python
soup.find('div',attrs={"class":"sell-point"}).text
```

```python
soup.find('span',attrs={"class":"off"}).text
```

```python
soup.find('span',attrs={"class":"price-num"}).text
```

```python
soup.find('span',attrs={"class":"old"}).text
```

```python
a=soup.find_all('h3',attrs={"class":"thick-font"})
Model_name=[i.text for i in a]
print(Model_name)
print(len(Model_name))
```

```python
b=soup.find_all('div',attrs={"class":"description"})
desc=[i.text for i in b]
print(desc)
print(len(desc))
```

```python
c=soup.find_all('span',attrs={"class":"off"})
offer=[i.text for i in c]
print(offer)
print(len(offer))
```

```python
d=soup.find_all('span',attrs={"class":"price-num"})
price=[i.text for i in d]
print(price)
print(len(price))
```

```python
e=soup.find_all('span',attrs={"class":"old"})
old_price=[i.text for i in e]
print(old_price)
print(len(old_price))
```

```python
f=soup.find_all('div',attrs={"class":"sell-point"})
sell_point=[i.text for i in f]
print(sell_point)
print(len(sell_point))
```

```python
import pandas as pd
```

```python
data=pd.DataFrame({"Model_Name":Model_name,"Memory_desc":desc,"discount":offer,"final_price":price,"Actual_price":old_price,"sell_point":sell_point})
data
```

```python
data.to_csv("Vivo_smart_phones.csv")
```
